```python
import numpy as np
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif

# Your estimator (copula-entropy MI)
#from MutualInformation import estimate_mi_from_ce  # :contentReference[oaicite:0]{index=0}

def bivariate_mi_mae_sklearn_vs_mind(
    X,
    y,
    *,
    # sklearn MI kwargs
    discrete_features=False,
    n_neighbors=3,
    random_state=0,
    # MIND / copula-entropy kwargs
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    task='regression'
):
    """
    Computes bivariate MI between each column X[:, j] and y using:
      (1) sklearn (KNN MI estimator)
      (2) your copula-entropy MI estimator (MIND-like via copula entropy)
    Returns MAE between the two MI vectors, plus both MI vectors.

    Notes:
      - If y looks discrete (integer/bool or low cardinality), uses mutual_info_classif.
      - Otherwise uses mutual_info_regression.
    """
    X = np.asarray(X)
    y = np.asarray(y)

    if X.ndim != 2:
        raise ValueError("X must be 2D (n_samples, n_features)")
    if y.ndim != 1:
        y = y.reshape(-1)
    if X.shape[0] != y.shape[0]:
        raise ValueError("X and y must have the same number of rows")

    n, d = X.shape

    # Decide sklearn estimator based on y "discreteness"
    y_is_intlike = np.issubdtype(y.dtype, np.integer) or (y.dtype == np.bool_)
    uniq = np.unique(y[~np.isnan(y)] if np.issubdtype(y.dtype, np.floating) else y).size
    y_is_low_card = uniq <= max(20, int(0.02 * n))  # heuristic


    if task=='classification':
        mi_sklearn = mutual_info_classif(
            X, y,
            discrete_features=discrete_features,
            n_neighbors=n_neighbors,
            random_state=random_state,
        ).astype(float)
    else:
        mi_sklearn = mutual_info_regression(
            X, y,
            discrete_features=discrete_features,
            n_neighbors=n_neighbors,
            random_state=random_state,
        ).astype(float)

    # Your MI per feature via copula entropy (1D X_j -> 1D y)
    mi_mind = np.empty(d, dtype=float)
    y2 = y.reshape(-1, 1)
    for j in range(d):
        xj = X[:, j].reshape(-1, 1)
        mi_mind[j] = estimate_mi_from_ce(
            xj, y2,
            degree=degree,
            max_num_mc=max_num_mc,
            max_num_mc_adapt=max_num_mc_adapt,
            alpha_reg=alpha_reg,
            random_state=random_state,
        )

    mae = float(np.mean(np.abs(mi_sklearn - mi_mind)))
    return {
        "mae": mae,
        "mi_sklearn": mi_sklearn,
        "mi_mind": mi_mind,
        "sklearn_mode": "classif" if task=='classification' else "regression",
    }
```

```python
import numpy as np

def mi_to_r2_bound(mi: float) -> float:
    mi = float(mi)
    if mi < 0:
        mi = 0.0
    return float(1.0 - np.exp(-2.0 * mi))

import numpy as np

def greedy_mi_feature_selection_adjscan_smart(
    X: np.ndarray,
    Y: np.ndarray,
    *,
    feature_names=None,
    max_features=None,
    r2_target=1,
    verbose=False,
    # passed through to estimate_mi_from_ce
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    random_state=None,
    feature_mode="hybrid_b",
    # permutation-adjustment
    n_perms: int = 50,
    perm_random_state=None,
    permute: str = "y",   # "y" (default) or "feature"
    null_q=0.5
):
    """
    Greedy forward selection with a permutation-adjusted monotone scan down the raw-MI ranking.

    Smart skip:
      - Let adj_prev be the current best adjusted score.
      - For a next candidate with raw MI = mi_raw:
            if mi_raw < adj_prev:
                it cannot beat adj_prev after adjustment (since adj <= raw and MI>=0),
                so we skip permutations for that candidate and continue scanning.
      - Otherwise compute adj_cur = mi_raw - mean(null), and compare:
            if adj_cur > adj_prev: update winner and keep scanning
            else stop and select the current winner.
    """
    X = np.asarray(X)
    Y = np.asarray(Y)

    if X.ndim != 2:
        raise ValueError("X must be 2D: (n, p)")
    if Y.ndim == 1:
        Y_ = Y.reshape(-1, 1)
    elif Y.ndim == 2:
        Y_ = Y
    else:
        raise ValueError("Y must be 1D or 2D")

    n, p = X.shape
    if Y_.shape[0] != n:
        raise ValueError("X and Y must have the same number of rows")

    if feature_names is None:
        feature_names = [f"x{j}" for j in range(p)]
    if len(feature_names) != p:
        raise ValueError("feature_names must have length p")

    if max_features is None:
        max_features = p
    max_features = int(max_features)

    rng_perm = np.random.default_rng(perm_random_state)

    def ok(diag):
      return (
          diag.get("reliable", False)
          and diag["ess_frac"] >= 0.02
          and diag["w_max"] <= 0.1
          and diag["theta_norm"] <= 50
          and diag["z_range"] <= 200
          and "nit=200" not in diag.get("stop_reason","")
      )

    def _mi(Xsub: np.ndarray, Ysub: np.ndarray) -> float:
        #mi, exy=estimate_mi_with_diag(Xsub,Ysub)
        #if not ok(exy):
        #  return -np.inf
        #return mi

        return estimate_mi_junction_tree_xy(Xsub,Ysub)
        return estimate_mi_block_factorization_xy(Xsub,Ysub)


        return float(
            estimate_mi_from_ce(
                Xsub,
                Ysub,
                degree=degree,
                max_num_mc=max_num_mc,
                max_num_mc_adapt=max_num_mc_adapt,
                alpha_reg=alpha_reg,
                random_state=random_state,
                feature_mode=feature_mode,
            )
        )

    def _null_q_for_candidate(Xsub: np.ndarray, Ysub: np.ndarray, q: float = 0.95) -> float:
      if n_perms is None or int(n_perms) <= 0:
          return float("nan")
      B = int(n_perms)
      null = np.empty(B, dtype=float)

      if permute.lower() == "y":
          for b in range(B):
              idx = rng_perm.permutation(n)
              null[b] = _mi(Xsub, Ysub[idx, :])
      elif permute.lower() == "feature":
          Xp = np.array(Xsub, copy=True)
          for b in range(B):
              idx = rng_perm.permutation(n)
              Xp[:, -1] = Xsub[idx, -1]
              null[b] = _mi(Xp, Ysub)
      else:
          raise ValueError("permute must be 'y' or 'feature'")

      return float(np.quantile(null, q))

    def _pick_by_adjscan_smart(X_sel: np.ndarray | None, remaining_idxs: list[int], step: int):
        # raw MI for all candidates
        cand_mi = np.empty(len(remaining_idxs), dtype=float)
        for t, j in enumerate(remaining_idxs):
            X_try = X[:, [j]] if X_sel is None else np.hstack([X_sel, X[:, [j]]])
            cand_mi[t] = _mi(X_try, Y_)

        order = np.argsort(-cand_mi)  # positions into remaining_idxs

        # seed with top-1: must compute its adjusted score to define adj_prev
        pos0 = int(order[0])
        j0 = int(remaining_idxs[pos0])
        X0 = X[:, [j0]] if X_sel is None else np.hstack([X_sel, X[:, [j0]]])
        nm0 = _null_q_for_candidate(X0, Y_,q=null_q)
        adj0 = float(cand_mi[pos0] - (0.0 if np.isnan(nm0) else nm0))
        if adj0 < 0.0:  # safety; MI theoretically nonneg, but estimator can wiggle
            adj0 = 0.0

        if verbose:
            print(f"Step {step}: raw MI ranking (top shown):")
            for k_show, pos in enumerate(order[: min(10, len(order))], start=1):
                j = remaining_idxs[int(pos)]
                print(f"  {k_show:>2d}. {feature_names[j]:>20s}  MI={cand_mi[int(pos)]:.6g}")
            if int(n_perms) > 0:
                print(f"  seed: {feature_names[j0]} adj={adj0:.6g} (null_mean={nm0:.6g})")
            else:
                print(f"  seed: {feature_names[j0]} (no perms)")
            print()

        j_prev, nm_prev, adj_prev = j0, float(nm0), float(adj0)

        # scan down
        for rank_i in range(1, len(order)):
            pos = int(order[rank_i])
            j = int(remaining_idxs[pos])
            mi_raw = float(cand_mi[pos])

            # smart skip: if even raw can't beat current adjusted, don't permute
            if mi_raw < adj_prev:
                if verbose and int(n_perms) > 0:
                    print(f"  skip perms: {feature_names[j]} raw={mi_raw:.6g} < prev_adj={adj_prev:.6g}")
                continue

            Xj = X[:, [j]] if X_sel is None else np.hstack([X_sel, X[:, [j]]])

            nm = _null_q_for_candidate(Xj, Y_, q=null_q)
            adj = mi_raw - (0.0 if np.isnan(nm) else float(nm))

            if adj < 0.0:
                adj = 0.0

            if verbose and int(n_perms) > 0:
                print(f"  compare: {feature_names[j]} adj={adj:.6g} (raw={mi_raw:.6g}, null_mean={nm:.6g}) "
                      f"vs prev_adj={adj_prev:.6g}")

            if adj > adj_prev:
                j_prev, nm_prev, adj_prev = j, float(nm), float(adj)
            else:
                # stop at first non-improvement (per your original scan logic)
                break

        j_pick = int(j_prev)
        # raw MI for picked candidate (needed to advance objective)
        # find its position:
        picked_pos = int(np.where(np.array(remaining_idxs, dtype=int) == j_pick)[0][0])
        mi_pick = float(cand_mi[picked_pos])

        if verbose:
            if int(n_perms) > 0:
                print(f"\n  chosen by smart-adjscan: {feature_names[j_pick]} "
                      f"adj={adj_prev:.6g} (raw={mi_pick:.6g}, null_mean={nm_prev:.6g})\n")
            else:
                print(f"\n  chosen by smart-adjscan: {feature_names[j_pick]} raw={mi_pick:.6g}\n")

        return j_pick, mi_pick, nm_prev, adj_prev, cand_mi

    remaining = list(range(p))
    selected = []
    history = []

    mi_cur = 0.0
    r2_cur = mi_to_r2_bound(mi_cur)

    step = 1
    while len(selected) < max_features and remaining:
        X_sel = None if len(selected) == 0 else X[:, selected]

        j_pick, mi_pick, nm_pick, adj_pick, cand_mi = _pick_by_adjscan_smart(X_sel, remaining, step)

        selected.append(j_pick)
        remaining.remove(j_pick)

        mi_base = mi_cur
        r2_base = r2_cur

        mi_adj = float(max(0.0, adj_pick))
        r2_adj = mi_to_r2_bound(mi_adj)

        mi_cur = mi_adj
        r2_cur = r2_adj

        history.append(dict(
            step=step,
            chosen_idx=j_pick,
            chosen_name=feature_names[j_pick],
            mi_raw=float(mi_pick),
            mi_adj=mi_adj,
            r2_bound=r2_adj,
            delta_mi=float(mi_adj - mi_base),
            delta_r2=float(r2_adj - r2_base),
            perm_null=float(nm_pick),
        ))

        if r2_target is not None and r2_cur >= float(r2_target):
            if verbose:
                print(f"Stopping: reached r2_target={float(r2_target):.6g} with R2_bound={r2_cur:.6g}.")
            break

        step += 1

    # convenience: univariate MI
    uni_mis = np.empty(p, dtype=float)
    for j in range(p):
        uni_mis[j] = _mi(X[:, [j]], Y_)

    return dict(
        selected_indices=selected,
        selected_names=[feature_names[i] for i in selected],
        history=history,
        univariate_mi=uni_mis,
    )

import numpy as np


def estimate_mi_from_ce(
    X,
    Y,
    *,
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0,
    random_state=None,
    feature_mode='pairwise',
    n_spline_basis=6,
    n_spline_degree=3,
    boundary_features=True,
    cross=True,
    pairwise=True,

):
    """
    MI estimator using copula entropy on a transformed feature bank:

        MI(X;Y) = Hc(T(X)) + Hc(T(Y)) - Hc([T(X), T(Y)])

    Assumes CopulaEntropyEstimator.compute() already performs
    the ECDF -> uniform transform internally.
    """

    X = np.asarray(X)
    Y = np.asarray(Y)

    if X.ndim == 1:
        X = X.reshape(-1, 1)
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)

    if X.shape[0] != Y.shape[0]:
        raise ValueError("X and Y must have same number of samples")

    def make_est():
        return CopulaEntropyEstimator(
            degree=degree,
            max_num_mc=max_num_mc,
            max_num_mc_adapt=max_num_mc_adapt,
            alpha_reg=alpha_reg,
            random_state=random_state,
            feature_mode=feature_mode,
            boundary_features=boundary_features,
            n_spline_degree=n_spline_degree,
            n_spline_basis=n_spline_basis,
            cross=cross,
            pairwise=pairwise
        )

    # Marginals
    h_x = 0.0 if X.shape[1] == 1 else float(make_est().compute(X)[0])
    #h_y = 0.0 if Y.shape[1] == 1 else float(make_est().compute(Y)[0])

    # Joint (must use the same transformed reps for the decomposition to make sense)
    XY = np.hstack([X, Y])
    h_xy = float(make_est().compute(XY)[0])

    return float(h_x - h_xy)
```

```python
import numpy as np
import pandas as pd
from scipy.special import gammaln, digamma
from numpy.linalg import slogdet, inv

# -----------------------------
# Multivariate Student-t sampler
# -----------------------------
def sample_mvt(n_samples, df, Omega, rng):
    p = Omega.shape[0]
    L = np.linalg.cholesky(Omega)
    z = rng.standard_normal((n_samples, p)) @ L.T
    g = rng.chisquare(df, size=n_samples) / df
    return z / np.sqrt(g)[:, None]

# -----------------------------
# Analytic MI for Student-t (pairwise or block)
# -----------------------------
def mi_student_t(n, m, df, Omega):
    Omega_xx = Omega[:n, :n]
    Omega_yy = Omega[n:n+m, n:n+m]

    sign, logdet_O = slogdet(Omega)
    signx, logdet_x = slogdet(Omega_xx)
    signy, logdet_y = slogdet(Omega_yy)
    if min(sign, signx, signy) <= 0:
        raise ValueError("Omega must be SPD.")

    I_gauss = 0.5 * (logdet_x + logdet_y - logdet_O)

    nu = df
    excess = (
        (gammaln(nu/2) + gammaln((nu+n+m)/2)
         - gammaln((nu+n)/2) - gammaln((nu+m)/2))
        - (nu+m)/2 * digamma((nu+m)/2)
        - (nu+n)/2 * digamma((nu+n)/2)
        + (nu+n+m)/2 * digamma((nu+n+m)/2)
        + nu/2 * digamma(nu/2)
    )
    return I_gauss + excess

# -----------------------------
# Control "how many unique ground-truth MIs"
# via how many unique rho(Xj,Y)
# -----------------------------
def make_rho_xy(n_features, n_unique_mi, low=0.05, high=0.85, values=None, shuffle=False, rng=None):
    """
    Returns rho_xy (len=n_features) with exactly n_unique_mi unique values.
    - If values is provided, uses those (len must be n_unique_mi).
    - Otherwise uses np.linspace(low, high, n_unique_mi).
    """
    if values is None:
        levels = np.linspace(low, high, n_unique_mi)
    else:
        levels = np.asarray(values, dtype=float)
        if len(levels) != n_unique_mi:
            raise ValueError("len(values) must equal n_unique_mi")
    reps = int(np.ceil(n_features / n_unique_mi))
    rho_xy = np.tile(levels, reps)[:n_features].copy()
    if shuffle:
        if rng is None:
            rng = np.random.default_rng(0)
        rng.shuffle(rho_xy)
    return rho_xy

def sigma_xx_equicorr(n_features, rho_xx):
    # Valid if rho_xx in (-1/(p-1), 1)
    p = n_features
    return (1 - rho_xx) * np.eye(p) + rho_xx * np.ones((p, p))

def enforce_spd_by_scaling_rho_xy(Sigma_xx, rho_xy, eps=1e-8):
    """
    Ensures rho_xy^T Sigma_xx^{-1} rho_xy < 1 by scaling rho_xy down if needed.
    Returns (rho_xy_scaled, scale_factor, quad_form_value_after).
    """
    Sinv = inv(Sigma_xx)
    q = float(rho_xy.T @ Sinv @ rho_xy)
    if q < 1 - eps:
        return rho_xy, 1.0, q
    # scale to target just below 1
    scale = np.sqrt((1 - eps) / q)
    rho2 = rho_xy * scale
    q2 = float(rho2.T @ Sinv @ rho2)
    return rho2, scale, q2

def omega_from_sigma_and_rho(Sigma_xx, rho_xy):
    p = Sigma_xx.shape[0]
    Omega = np.zeros((p + 1, p + 1), dtype=float)
    Omega[:p, :p] = Sigma_xx
    Omega[:p, -1] = rho_xy
    Omega[-1, :p] = rho_xy
    Omega[-1, -1] = 1.0
    return Omega

# -----------------------------
# Per-dataset evaluation (feature-wise MI vs Y)
# -----------------------------
def evaluate_dataset_joint(X, Y, df, Omega, dataset_meta):
    p = X.shape[1]
    true_mi = mi_student_t(p, 1, df, Omega)

    est_block = estimate_mi_block_factorization_xy(X, Y)
    est_jtree = estimate_mi_junction_tree_xy(X, Y)
    est_ce    = estimate_mi_from_ce(X, Y)

    return pd.DataFrame([{
        **dataset_meta,
        "true_mi": float(true_mi),
        "block": float(est_block),
        "junction_tree": float(est_jtree),
        "copula_entropy": float(est_ce),
    }])


import numpy as np
from numpy.linalg import inv

def rho_xy_for_target_q(Sigma_xx, q_target, rng, max_abs_rho=0.95, max_tries=10_000):
    """
    Construct rho_xy such that rho_xy.T @ inv(Sigma_xx) @ rho_xy == q_target,
    while keeping |rho_i| <= max_abs_rho.

    q_target must satisfy 0 < q_target < 1.
    """
    if not (0.0 < q_target < 1.0):
        raise ValueError("q_target must be in (0,1).")

    Sinv = inv(Sigma_xx)

    for _ in range(max_tries):
        # random direction
        v = rng.standard_normal(Sigma_xx.shape[0])

        # compute current quadratic form
        q0 = float(v.T @ Sinv @ v)
        if q0 <= 0:
            continue

        # scale to match target q
        s = np.sqrt(q_target / q0)
        rho = s * v

        # keep correlations reasonable
        if np.max(np.abs(rho)) <= max_abs_rho:
            # verify (numerically)
            q = float(rho.T @ Sinv @ rho)
            return rho, q

    raise RuntimeError("Failed to sample rho_xy within bounds; relax max_abs_rho or q_target.")


def make_q_levels(n_unique, q_low=0.02, q_high=0.98):
    # q in (0,1). Don't go too close to 1 unless you want huge MI.
    if not (0.0 < q_low < q_high < 1.0):
        raise ValueError("Require 0 < q_low < q_high < 1.")
    return np.linspace(q_low, q_high, n_unique)


def q_from_target_total_mi(I_target, p, df):
    """
    Convert total MI target to q using:
      I_total = I_gauss + excess
      I_gauss = -0.5 * log(1-q)
    """
    # excess term is constant for given (p,1,df)
    # We can get it by evaluating mi_student_t with Omega = blockdiag(I_p, 1)
    Omega_ind = np.eye(p + 1)
    excess = mi_student_t(p, 1, df, Omega_ind)  # since I_gauss=0 when independent
    I_gauss = I_target - excess
    if I_gauss <= 0:
        raise ValueError(f"Target MI too small; must exceed excess={excess:.6f} nats for (p={p},df={df}).")
    q = 1.0 - np.exp(-2.0 * I_gauss)
    if not (0.0 < q < 1.0):
        raise ValueError("Converted q not in (0,1).")
    return float(q)

def benchmark_unique_joint_mi(
    *,
    N=3000,
    n_features=5,
    df=4,
    n_unique_mi=5,
    q_low=0.02,
    q_high=0.98,
    rho_xx=0.2,
    n_datasets=10,
    seed0=0,
    max_abs_rho=0.95,
):
    frames = []
    run_id = 0

    # fixed Sigma_xx across datasets (you can also vary it if you want)
    Sigma_xx = sigma_xx_equicorr(n_features, rho_xx)

    # choose q-levels that imply unique analytic MI
    q_levels = make_q_levels(n_unique_mi, q_low=q_low, q_high=q_high)

    for k in range(n_datasets):
        rng = np.random.default_rng(seed0 + k)

        # cycle through q-levels so you actually realize all unique values
        q_target = float(q_levels[k % n_unique_mi])

        # build rho_xy to achieve q_target
        rho_xy, q_actual = rho_xy_for_target_q(
            Sigma_xx, q_target, rng, max_abs_rho=max_abs_rho
        )

        Omega = omega_from_sigma_and_rho(Sigma_xx, rho_xy)

        # sample
        Z = sample_mvt(N, df, Omega, rng)
        X = Z[:, :n_features]
        Y = Z[:, -1]

        meta = {
            "run_id": run_id,
            "N": N,
            "n_features": n_features,
            "df": df,
            "rho_xx": rho_xx,
            "q_target": q_target,
            "q_actual": float(q_actual),
            "seed": int(seed0 + k),
        }

        frames.append(evaluate_dataset_joint(X, Y, df, Omega, meta))
        run_id += 1

    out = pd.concat(frames, ignore_index=True)

    # sanity: how many unique ground-truth values did we get?
    # (rounded to avoid tiny numeric jitter)
    out["true_mi_rounded"] = out["true_mi"].round(10)
    print("Unique true MI:", out["true_mi_rounded"].nunique())

    return out

def compute_mae(df):
    return pd.Series({
        "block_mae": np.mean(np.abs(df["block"] - df["true_mi"])),
        "junction_tree_mae": np.mean(np.abs(df["junction_tree"] - df["true_mi"])),
        "copula_entropy_mae": np.mean(np.abs(df["copula_entropy"] - df["true_mi"])),
    })

# -----------------------------
# Example usage
# -----------------------------
joint_df = benchmark_unique_joint_mi(
    N=3_000,
    n_features=5,
    df=5,
    n_unique_mi=30,
    q_low=0.02,
    q_high=0.98,
    rho_xx=0.25,
    n_datasets=90,   # make sure >= n_unique_mi if you want to see them all
    seed0=123
)

print(joint_df[["true_mi","block","junction_tree","copula_entropy"]].corr(method="spearman"))
print(compute_mae(joint_df))
```

```python
import numpy as np
import pandas as pd

# -----------------------------
# Utility: Spearman + bootstrap CI (over rows)
# -----------------------------
def _rankdata_1d(a):
    # fast average-rank for ties using pandas
    return pd.Series(a).rank(method="average").to_numpy()

def spearman_rho(x, y):
    rx = _rankdata_1d(x)
    ry = _rankdata_1d(y)
    # Pearson on ranks
    rx = rx - rx.mean()
    ry = ry - ry.mean()
    denom = np.sqrt((rx**2).sum() * (ry**2).sum())
    return 0.0 if denom == 0 else float((rx * ry).sum() / denom)

def bootstrap_ci_spearman(df, xcol, ycol, n_boot=300, seed=0):
    rng = np.random.default_rng(seed)
    x = df[xcol].to_numpy()
    y = df[ycol].to_numpy()
    n = len(df)

    base = spearman_rho(x, y)
    if n <= 3:
        return base, np.nan, np.nan

    boots = np.empty(n_boot, dtype=float)
    for b in range(n_boot):
        idx = rng.integers(0, n, size=n)
        boots[b] = spearman_rho(x[idx], y[idx])

    lo, hi = np.quantile(boots, [0.025, 0.975])
    return base, float(lo), float(hi)


# -----------------------------
# Core: run one p, return per-dataset rows
# (uses your benchmark_unique_joint_mi but "p" is n_features)
# -----------------------------
def run_one_p(
    p,
    *,
    N=2000,
    df=5,
    rho_xx=0.25,
    n_unique_mi=25,
    q_low=0.02,
    q_high=0.98,
    n_datasets=60,
    seed0=123,
    max_abs_rho=0.95,
):
    out = benchmark_unique_joint_mi(
        N=N,
        n_features=p,
        df=df,
        n_unique_mi=n_unique_mi,
        q_low=q_low,
        q_high=q_high,
        rho_xx=rho_xx,
        n_datasets=n_datasets,
        seed0=seed0,
        max_abs_rho=max_abs_rho,
    ).copy()
    out["p"] = p
    return out


# -----------------------------
# Sweep over p and summarize Spearman per estimator
# -----------------------------
def sweep_features_spearman(
    p_grid,
    *,
    N=2000,
    df=5,
    rho_xx=0.25,
    n_unique_mi=25,
    q_low=0.02,
    q_high=0.98,
    n_datasets=60,
    seed0=123,
    max_abs_rho=0.95,
    n_boot=300,
):
    all_rows = []
    summaries = []

    for i, p in enumerate(p_grid):
        joint_df = run_one_p(
            p,
            N=N,
            df=df,
            rho_xx=rho_xx,
            n_unique_mi=n_unique_mi,
            q_low=q_low,
            q_high=q_high,
            n_datasets=n_datasets,
            seed0=seed0 + 10_000*i,   # decorrelate across p
            max_abs_rho=max_abs_rho,
        )
        all_rows.append(joint_df)

        for method in ["block", "junction_tree", "copula_entropy"]:
            rho, lo, hi = bootstrap_ci_spearman(
                joint_df, "true_mi", method, n_boot=n_boot, seed=seed0 + 999*i
            )
            summaries.append({
                "p": p,
                "method": method,
                "spearman": rho,
                "ci_lo": lo,
                "ci_hi": hi,
                "N": N,
                "n_datasets": n_datasets,
                "n_unique_mi": n_unique_mi,
                "df": df,
                "rho_xx": rho_xx,
            })

    all_rows = pd.concat(all_rows, ignore_index=True)
    summary_df = pd.DataFrame(summaries).sort_values(["method", "p"]).reset_index(drop=True)
    return all_rows, summary_df


# -----------------------------
# Example: "fast but minimally robust"
# -----------------------------
p_grid = [2, 3, 5, 7,9]#, 12, 16, 24, 32]

all_rows, summary_df = sweep_features_spearman(
    p_grid,
    N=2000,           # keep moderate
    df=5,
    rho_xx=0.25,
    n_unique_mi=30,   # enough distinct ground-truth ordering for Spearman
    n_datasets=30,    # enough to stabilize rank correlation
    q_low=0.05,       # avoid extreme near-0 MI
    q_high=0.95,      # avoid near-singularity / huge MI
    n_boot=300,       # cheap bootstrap over datasets
    seed0=123,
)

print(summary_df.pivot(index="p", columns="method", values="spearman"))

# If you want the raw per-p correlation matrices too:
for p in p_grid:
    sub = all_rows[all_rows["p"] == p]
    print(
    abs(sub['true_mi']-sub['block']).mean(),
    abs(sub['true_mi']-sub['copula_entropy']).mean(),
    abs(sub['true_mi']-sub['junction_tree']).mean())
    corr = sub[["true_mi", "block", "junction_tree", "copula_entropy"]].corr(method="spearman")
    print(f"\n=== p={p} Spearman matrix ===")
    print(corr)
```

```python
import numpy as np
from scipy.stats import spearmanr

# If you have this already, keep yours
def _max_spanning_tree_parents(W: np.ndarray, root: int = 0):
    d = W.shape[0]
    in_tree = np.zeros(d, dtype=bool)
    pa = -np.ones(d, dtype=int)
    best = -np.inf * np.ones(d, dtype=float)

    in_tree[root] = True
    best[:] = W[root]
    pa[:] = root
    pa[root] = -1
    best[root] = -np.inf

    for _ in range(d - 1):
        v = int(np.argmax(np.where(in_tree, -np.inf, best)))
        in_tree[v] = True
        for u in range(d):
            if not in_tree[u] and W[v, u] > best[u]:
                best[u] = W[v, u]
                pa[u] = v
    return pa

def _blocks_from_mst(
    pa: np.ndarray,
    W: np.ndarray,
    *,
    min_edge_weight: float = 0.0,
    max_block_size: int | None = None,
):
    """
    Turn MST into blocks by:
      1) removing edges with weight < min_edge_weight
      2) additionally removing weakest remaining edges until all components <= max_block_size (if set)
    """
    d = pa.shape[0]
    edges = []
    for i in range(d):
        p = int(pa[i])
        if p >= 0:
            edges.append((i, p, float(W[i, p])))

    # Start with all edges, then mark removals
    kept = np.ones(len(edges), dtype=bool)
    for eidx, (_, _, w) in enumerate(edges):
        if w < float(min_edge_weight):
            kept[eidx] = False

    def components_from_kept():
        adj = [[] for _ in range(d)]
        for (i, p, _), k in zip(edges, kept):
            if k:
                adj[i].append(p)
                adj[p].append(i)
        seen = np.zeros(d, dtype=bool)
        comps = []
        for s in range(d):
            if seen[s]:
                continue
            stack = [s]
            seen[s] = True
            comp = []
            while stack:
                v = stack.pop()
                comp.append(v)
                for u in adj[v]:
                    if not seen[u]:
                        seen[u] = True
                        stack.append(u)
            comps.append(comp)
        return comps

    comps = components_from_kept()

    if max_block_size is not None:
        max_block_size = int(max_block_size)
        if max_block_size < 1:
            raise ValueError("max_block_size must be >= 1")

        # Iteratively cut weakest edge inside any oversized component
        while True:
            comps = components_from_kept()
            oversized = [c for c in comps if len(c) > max_block_size]
            if not oversized:
                break

            # pick one oversized component; cut its weakest kept internal edge
            comp_set = set(oversized[0])
            internal = []
            for eidx, (i, p, w) in enumerate(edges):
                if kept[eidx] and (i in comp_set) and (p in comp_set):
                    internal.append((w, eidx))
            if not internal:
                # can't split further with remaining MST edges (shouldn't happen)
                break
            _, eidx_min = min(internal, key=lambda t: t[0])
            kept[eidx_min] = False

    comps = components_from_kept()
    blocks = [np.array(sorted(c), dtype=int) for c in comps]
    return blocks, edges, kept

def estimate_mi_block_factorization_xy(
    X: np.ndarray,
    Y: np.ndarray,
    *,
    dep_mode: str = "mi",   # "spearman" (fast) or "mi"
    root: int = 0,
    min_edge_weight: float = 0.0,
    max_block_size: int | None = 6,

    # --- estimators ---
    # Used only if dep_mode="mi" (for structure learning on X pairs)
    mi_pair_estimator=None,  # (xi, xj) -> float MI, 1D/2D ok

    # Copula entropy estimator for multivariate Hc(Z).
    # If None, we'll build one from CopulaEntropyEstimator below.
    ce_estimator=None,  # callable: (Z: (n,k)) -> float Hc(Z)

    # pass-through knobs (match your CopulaEntropyEstimator / wrapper)
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    random_state=None,
):
    """
    Block-factorized analogue of your junction-tree estimator.

    Structure:
      - Learn an MST on X using dep_mode
      - Cut MST into blocks (connected components) via min_edge_weight and/or max_block_size
      - Assume a star junction tree around Y with cliques (Y, X_block) and separator Y,
        AND assume X copula factorizes across blocks (so Hc(X) ≈ Σ Hc(X_block)).

    Returns:
      mi_hat, diagnostics
    """
    X = np.asarray(X)
    Y = np.asarray(Y)
    if X.ndim != 2:
        raise ValueError("X must be 2D")
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)
    elif Y.ndim != 2:
        raise ValueError("Y must be 1D or 2D")

    n, d = X.shape
    if Y.shape[0] != n:
        raise ValueError("X and Y must have same n")

    # --- build ce_estimator if not provided ---
    if ce_estimator is None:
        # expects your local class from CopulaEntropy.py
        from CopulaEntropy import CopulaEntropyEstimator

        ce = CopulaEntropyEstimator(
            degree=degree,
            max_num_mc=max_num_mc,
            max_num_mc_adapt=max_num_mc_adapt,
            alpha_reg=alpha_reg,
            random_state=random_state,
        )

        def ce_estimator(Z):

            h, _ = ce.compute(np.asarray(Z))
            return float(h)
    else:
        ce = None  # we won't have per-call diagnostics unless user wraps it

    # --- dependency matrix on X ---
    W = np.zeros((d, d), dtype=float)
    if dep_mode.lower() == "spearman":
        for i in range(d):
            for j in range(i + 1, d):
                rho = spearmanr(X[:, i], X[:, j]).correlation
                w = 0.0 if rho is None or np.isnan(rho) else abs(float(rho))
                W[i, j] = W[j, i] = w
    elif dep_mode.lower() == "mi":
        if mi_pair_estimator is None:
            # If you didn't pass a MI estimator, use the fact that for 1D pairs:
            #   MI(xi;xj) = -Hc([xi,xj])
            for i in range(d):
                for j in range(i + 1, d):
                    hc_ij = ce_estimator(np.column_stack([X[:, i], X[:, j]]))
                    W[i, j] = W[j, i] = float(-hc_ij)
        else:
            for i in range(d):
                xi = X[:, i]
                for j in range(i + 1, d):
                    xj = X[:, j]
                    W[i, j] = W[j, i] = float(mi_pair_estimator(
                        xi, xj,
                        degree=degree,
                        max_num_mc=max_num_mc,
                        max_num_mc_adapt=max_num_mc_adapt,
                        alpha_reg=alpha_reg,
                        random_state=random_state,
                    ))
    else:
        raise ValueError("dep_mode must be 'spearman' or 'mi'")

    # --- MST on X and convert to blocks ---
    pa = _max_spanning_tree_parents(W, root=root)
    blocks, mst_edges, kept_mask = _blocks_from_mst(
        pa, W,
        min_edge_weight=min_edge_weight,
        max_block_size=max_block_size,
    )

    # --- copula entropy terms (direct, multivariate) ---
    def Hc(Z):
        Z = np.asarray(Z)
        if Z.ndim == 1:
            Z = Z.reshape(-1, 1)
        return float(ce_estimator(Z))

    hc_Y = Hc(Y)  # for 1D Y this should be ~0

    hc_X = 0.0
    hc_XY = 0.0
    per_block = []

    for b_idx, idxs in enumerate(blocks):
        Xb = X[:, idxs]
        hc_Xb = Hc(Xb)
        hc_X += hc_Xb

        Z = np.concatenate([Y, Xb], axis=1)
        hc_YXb = Hc(Z)

        hc_XY += hc_YXb
        per_block.append(
            dict(
                block=b_idx,
                idxs=idxs,
                size=int(len(idxs)),
                hc_Xb=float(hc_Xb),
                hc_YXb=float(hc_YXb),
                mi_block=float(hc_Xb + hc_Y - hc_YXb),  # equals MI(Xb;Y) if estimator is exact
            )
        )

    B = len(blocks)
    # Star junction-tree correction for the shared separator Y
    hc_XY = hc_XY - float(B - 1) * hc_Y

    mi_hat = float(hc_X + hc_Y - hc_XY)
    if mi_hat < 0:
        mi_hat = 0.0

    diag = {
        "parents": pa,
        "blocks": blocks,
        "num_blocks": B,
        "hc_X": float(hc_X),
        "hc_Y": float(hc_Y),
        "hc_XY": float(hc_XY),
        "per_block": per_block,
        "mst_edges": mst_edges,
        "mst_kept_mask": kept_mask,
    }
    if ce is not None:
        diag["ce_last_diag"] = dict(getattr(ce, "last_diag", {}) or {})
        diag["ce_last_reliable"] = bool(getattr(ce, "last_reliable", True))

    return mi_hat#, diag
```

```python
import numpy as np
from scipy.stats import spearmanr

def _max_spanning_tree_parents(W: np.ndarray, root: int = 0):
    """
    Prim over dense weight matrix W (dxd), returns parent array pa with pa[root]=-1.
    Assumes larger W => stronger dependency.
    """
    d = W.shape[0]
    in_tree = np.zeros(d, dtype=bool)
    pa = -np.ones(d, dtype=int)
    best = -np.inf * np.ones(d, dtype=float)

    in_tree[root] = True
    best[:] = W[root]
    pa[:] = root
    pa[root] = -1
    best[root] = -np.inf

    for _ in range(d - 1):
        v = int(np.argmax(np.where(in_tree, -np.inf, best)))
        in_tree[v] = True
        # relax edges from v
        for u in range(d):
            if not in_tree[u] and W[v, u] > best[u]:
                best[u] = W[v, u]
                pa[u] = v

    return pa

def estimate_mi_junction_tree_xy(
    X: np.ndarray,
    y: np.ndarray,
    *,
    dep_mode: str = "spearman",      # "mi" or "spearman"
    root: int = 0,
    # you already have this:
    mi_estimator=estimate_mi_from_ce,         # function (A,B)-> float; must accept 1D or 2D arrays
    # pass-through knobs for your estimator if you want:
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    random_state=None,
):
    """
    MI(X; y) using a treewidth-2 junction tree:
      - build max spanning tree on X using dep_mode
      - compute Hc(X) from X-tree edges (2D blocks)
      - compute Hc(X,y) from cliques {y, Xi, Xpa(i)} (3D blocks) and separators {y, Xpa(i)} (2D blocks)

    Requires only bivariate MI (or Spearman) to learn structure, and
    only 2D/3D copula-entropy blocks for estimation.
    """
    if mi_estimator is None:
        raise ValueError("Provide mi_estimator, e.g. your estimate_mi_from_ce.")

    #X = np.asarray(X)
    #y = np.asarray(y).reshape(-1, 1)

    if X.ndim != 2:
        raise ValueError("X must be 2D")
    n, d = X.shape
    #if y.shape[0] != n:
     #   raise ValueError("X and y must have same n")


    # ---- 1) build dependency matrix on X ----
    W = np.zeros((d, d), dtype=float)
    if dep_mode.lower() == "spearman":
        for i in range(d):
            for j in range(i + 1, d):
                rho = spearmanr(X[:, i], X[:, j]).correlation
                w = 0.0 if rho is None or np.isnan(rho) else abs(float(rho))
                W[i, j] = W[j, i] = w
    elif dep_mode.lower() == "mi":
        for i in range(d):
            xi = X[:, i]
            for j in range(i + 1, d):
                xj = X[:, j]
                w = float(mi_estimator(
                    xi, xj,
                    degree=degree,
                    max_num_mc=max_num_mc,
                    max_num_mc_adapt=max_num_mc_adapt,
                    alpha_reg=alpha_reg,
                    random_state=random_state,
                ))
                W[i, j] = W[j, i] = w
    else:
        raise ValueError("dep_mode must be 'mi' or 'spearman'")

    # ---- 2) max spanning tree on X ----
    pa = _max_spanning_tree_parents(W, root=root)

    # ---- 3) helper: copula entropy from MI estimator ----
    # For continuous vars: MI(A;B) = Hc(A) + Hc(B) - Hc([A,B])
    # and Hc(1D)=0 => Hc([A,B]) = -MI(A;B) for 1D A,B.
    def Hc_2(a, b):
        # returns copula entropy of the 2D pair (a,b)
        mi = float(mi_estimator(
            np.asarray(a).reshape(n, 1),
            np.asarray(b).reshape(n, 1),
            degree=degree,
            max_num_mc=max_num_mc,
            max_num_mc_adapt=max_num_mc_adapt,
            alpha_reg=alpha_reg,
            random_state=random_state,
        ))
        return -mi

    # For 3 vars: we can estimate Hc(Y, Xi, Xj) by directly calling CE estimator on 3D,
    # but you said you want to use the "2d X to 1d y" building block. Your estimate_mi_from_ce
    # returns MI between (Xi,Xj) and y if passed X as 2D and y as 1D.
    #
    # We need Hc([y, Xi, Xj]), not MI((Xi,Xj);y). Use:
    #   MI((Xi,Xj);y) = Hc(Xi,Xj) + Hc(y) - Hc(y, Xi, Xj)
    # -> Hc(y, Xi, Xj) = Hc(Xi,Xj) - MI((Xi,Xj);y)   (since Hc(y)=0)
    def Hc_3_y_xx(i, j):
        mi_xx_y = float(mi_estimator(
            X[:, [i, j]],
            y,
            degree=degree,
            max_num_mc=max_num_mc,
            max_num_mc_adapt=max_num_mc_adapt,
            alpha_reg=alpha_reg,
            random_state=random_state,
        ))
        hc_xx = Hc_2(X[:, i], X[:, j])
        return hc_xx - mi_xx_y

    # ---- 4) compute Hc(X) from X-tree ----
    hc_X = 0.0
    for i in range(d):
        p = pa[i]
        if p >= 0:
            hc_X += Hc_2(X[:, i], X[:, p])

    # ---- 5) compute Hc(X,y) from junction tree ----
    # Hc(X,y) ≈ Hc(y, X_root) + Σ_{i!=root} Hc(y, Xi, Xpa(i)) - Σ_{i!=root} Hc(y, Xpa(i))
    hc_y_root = Hc_2(y[:, 0], X[:, root])

    hc_cliques = 0.0
    hc_seps = 0.0
    for i in range(d):
        p = pa[i]
        if p >= 0:
            hc_cliques += Hc_3_y_xx(i, p)      # clique {y, Xi, Xp}
            hc_seps += Hc_2(y[:, 0], X[:, p])  # separator {y, Xp}

    hc_Xy = hc_y_root + hc_cliques - hc_seps

    # ---- 6) MI estimate ----
    mi_xy = hc_X - hc_Xy
    if mi_xy < 0:
        mi_xy = 0.0
    return float(mi_xy)#, {"parents": pa, "hc_X": hc_X, "hc_Xy": hc_Xy}
```

```python
from utils import _modified_ecdf_to_uniform
import numpy as np

def jt_separator_forward_select(
    X,
    y,
    *,
    max_features=9099,
    # JT structure learning (pairwise only)
    jt_max_dim=2,
    jt_min_dim=2,
    dep_metric="mi",
    dep_top_k=None,
    undersized="merge",
    # Incremental MI scoring
    sep_cap=6,              # cap |Sep(j) ∩ selected| using pairwise weights
    # --- passed to your copula-entropy MI estimator ---
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    random_state=None,
    feature_mode="pairwise",
    n_spline_basis=6,
    n_spline_degree=3,
    boundary_features=False,
    cross=False,
    pairwise=False,
    # Stopping
    n_perm=1,
    stop_if_leq_perm_mean=False,
    min_gain=0.0,
    # OPTIONAL: structural heuristics (1) + (2)
    use_leaf_penalty=False,
    leaf_penalty=0.0,
    use_glue_bonus=False,
    glue_bonus=0.0,
):
    """
    Forward selection using:
      - early bivariate MI to seed the first feature
      - a fixed junction-tree structure from pairwise dependence on X
      - per-candidate scoring: I(y; x_j | C(j)), where C(j) = Sep(j) ∩ selected (capped)

    Candidate scoring uses your copula-entropy MI estimator:
        estimate_mi_from_ce(X_subset, y, ...)

    Notes
    -----
    - This assumes your estimate_mi_from_ce already does the ECDF->uniform internally.
    - For JT structure learning on X, we use the same rank/(n+1) mapping as your utils
      (_modified_ecdf_to_uniform). :contentReference[oaicite:0]{index=0}
    """
    X = np.asarray(X)
    y = np.asarray(y)
    if y.ndim == 1:
        y = y.reshape(-1, 1)

    n, d = X.shape
    if max_features is None:
        max_features = d
    max_features = int(min(max_features, d))

    rng = np.random.default_rng(random_state)

    # ----------------------------
    # Helpers
    # ----------------------------
    def _rank_to_unit_interval(A: np.ndarray) -> np.ndarray:
        # matches your framework: column-wise stable ranks / (n+1)
        return _modified_ecdf_to_uniform(np.asarray(A))

    def _mi_xy(cols):
        cols = tuple(cols)
        return float(
            estimate_mi_junction_tree_xy(
                X[:, cols],
                y,
            )
        )

    def _cond_gain(j, C):
        C = tuple(C)
        if len(C) == 0:
            return _mi_xy((j,))

        base = mi_cache_yC.get(C)
        if base is None:
            base = _mi_xy(C)
            mi_cache_yC[C] = base

        Cj = tuple(sorted(C + (j,)))
        full = mi_cache_yC.get(Cj)
        if full is None:
            full = _mi_xy(Cj)
            mi_cache_yC[Cj] = full

        return max(float(full - base), 0.0)

    # ----------------------------
    # Seed: bivariate MI ranking
    # ----------------------------
    mi_cache_yC = {}  # cache I(y; subset) for subset tuples (subset is X-cols)
    mi_biv = np.empty(d, dtype=float)
    for j in range(d):
        mi_biv[j] = _mi_xy((j,))
    first = int(np.argmax(mi_biv))

    selected = [first]
    gains = [float(mi_biv[first])]  # adjusted later if heuristics enabled

    # ----------------------------
    # Learn JT structure on X (pairwise only)
    # ----------------------------
    Ux = _rank_to_unit_interval(X)
    W = _pairwise_dependence_graph(Ux, metric=dep_metric, top_k=dep_top_k)

    cliques, jt_edges = learn_jt_structure_from_samples(
        Ux,
        max_dim=jt_max_dim,
        min_dim=jt_min_dim,
        dep_metric=dep_metric,
        dep_top_k=dep_top_k,
        undersized=undersized,
    )

    # Build clique adjacency (tree) with separators
    m = len(cliques)
    adj = [[] for _ in range(m)]  # (nbr, sep_tuple)
    for i, j, sep in jt_edges:
        adj[i].append((j, tuple(sep)))
        adj[j].append((i, tuple(sep)))

    # Root the clique-tree at a clique containing the first selected feature
    def _best_root_for_feature(f):
        candidates = [i for i, C in enumerate(cliques) if f in C]
        if not candidates:
            return 0
        best = None
        best_key = None
        for i in candidates:
            C = cliques[i]
            key = (len(C), float(np.sum([W[f, u] for u in C if u != f])))
            if best_key is None or key > best_key:
                best_key = key
                best = i
        return int(best)

    root = _best_root_for_feature(first)

    parent = np.full(m, -1, dtype=int)
    sep_to_parent = [tuple() for _ in range(m)]

    stack = [root]
    parent[root] = root
    while stack:
        i = stack.pop()
        for j, sep in adj[i]:
            if parent[j] != -1:
                continue
            parent[j] = i
            sep_to_parent[j] = tuple(sep)
            stack.append(j)

    # Assign each variable to a "home clique" (max total dep to members)
    var_home = np.full(d, -1, dtype=int)
    for v in range(d):
        containing = [i for i, C in enumerate(cliques) if v in C]
        if not containing:
            var_home[v] = root
            continue
        best = None
        best_score = None
        for i in containing:
            C = cliques[i]
            sc = float(np.sum([W[v, u] for u in C if u != v]))
            key = (len(C), sc)
            if best_score is None or key > best_score:
                best_score = key
                best = i
        var_home[v] = int(best)

    # Sep(var) = separator on edge (home_clique -> parent)
    SepVar = [tuple() for _ in range(d)]
    for v in range(d):
        hc = var_home[v]
        SepVar[v] = tuple() if hc == root else tuple(sep_to_parent[hc])

    # Reverse index: which candidates are affected when we add s to selected?
    rev = [[] for _ in range(d)]
    for j in range(d):
        for s in SepVar[j]:
            s = int(s)
            if 0 <= s < d:
                rev[s].append(j)

    # ----------------------------
    # OPTIONAL: (1) leaf-only + (2) glue counts
    # ----------------------------
    leaf_only = None
    sep_count = None
    sep_norm = None

    if use_leaf_penalty or use_glue_bonus:
        clique_deg = np.zeros(m, dtype=int)
        sep_count = np.zeros(d, dtype=int)

        for i, j, S in jt_edges:
            clique_deg[i] += 1
            clique_deg[j] += 1
            for f in S:
                f = int(f)
                if 0 <= f < d:
                    sep_count[f] += 1

        is_leaf_clique = (clique_deg == 1)
        if m == 1:
            is_leaf_clique[:] = False

        if use_leaf_penalty:
            feat_clique_count = np.zeros(d, dtype=int)
            feat_only_clique = np.full(d, -1, dtype=int)
            for ci, C in enumerate(cliques):
                for f in C:
                    f = int(f)
                    if 0 <= f < d:
                        feat_clique_count[f] += 1
                        feat_only_clique[f] = ci

            leaf_only = np.zeros(d, dtype=bool)
            for f in range(d):
                if feat_clique_count[f] == 1:
                    ci = feat_only_clique[f]
                    if ci >= 0 and is_leaf_clique[ci]:
                        leaf_only[f] = True

        if use_glue_bonus:
            max_sep = int(np.max(sep_count)) if d else 0
            sep_norm = (sep_count / max_sep) if max_sep > 0 else np.zeros(d, dtype=float)

    def _adjust(j, base):
        s = float(base)
        if use_leaf_penalty and leaf_penalty and leaf_only is not None and leaf_only[j]:
            s -= float(leaf_penalty)
        if use_glue_bonus and glue_bonus and sep_count is not None and sep_count[j] > 0:
            s += float(glue_bonus) * float(sep_norm[j])
        return s

    # If heuristics are on, adjust the first gain (for reporting consistency)
    gains[0] = _adjust(first, gains[0])

    # ----------------------------
    # Iterative selection with cached + sparse updates
    # ----------------------------
    selected_set = set(selected)
    cached_C = [None] * d
    cached_score = np.zeros(d, dtype=float)
    active = np.ones(d, dtype=bool)
    active[first] = False

    def _make_C(j):
        S = [s for s in SepVar[j] if s in selected_set]
        if len(S) <= sep_cap:
            return tuple(sorted(S))
        S.sort(key=lambda s: W[j, s], reverse=True)
        return tuple(sorted(S[:sep_cap]))

    # initial scoring of all candidates (once)
    for j in range(d):
        if not active[j]:
            continue
        Cj = _make_C(j)
        cached_C[j] = Cj
        base = _cond_gain(j, Cj)
        cached_score[j] = _adjust(j, base)

    while len(selected) < max_features:
        j_star = int(np.argmax(np.where(active, cached_score, -np.inf)))
        best_adj = float(cached_score[j_star])

        if not np.isfinite(best_adj) or best_adj <= float(min_gain):
            break

        # permutation-based stop test for the winner, using its current C(j*)
        # use UNADJUSTED gain for the test (heuristics shouldn't "beat" permutations)
        Cstar = cached_C[j_star]
        if n_perm and stop_if_leq_perm_mean:
            best_unadj = float(_cond_gain(j_star, Cstar))
            perm_gains = []
            xj = X[:, j_star].copy()
            for _ in range(int(n_perm)):
                X[:, j_star] = rng.permutation(xj)
                perm_gains.append(_cond_gain(j_star, Cstar))
            X[:, j_star] = xj
            if best_unadj <= float(np.mean(perm_gains)):
                break

        # accept
        selected.append(j_star)
        gains.append(best_adj)
        selected_set.add(j_star)
        active[j_star] = False

        # sparse update: only candidates whose separator includes j_star can change C(j)
        for j in rev[j_star]:
            if not active[j]:
                continue
            newC = _make_C(j)
            if newC == cached_C[j]:
                continue
            cached_C[j] = newC
            base = _cond_gain(j, newC)
            cached_score[j] = _adjust(j, base)

    return selected, gains, cached_score
```

```python
import numpy as np
from utils import _modified_ecdf_to_uniform

def bf_separator_forward_select(
    X,
    y,
    *,
    max_features=9099,

    # block learning on X
    dep_metric="mi",          # "mi" or "spearman" if your W builder supports it
    dep_top_k=None,
    root=0,
    min_edge_weight=0.0,
    max_block_size=6,

    # conditional neighborhood
    sep_cap=6,                # cap |C(j)| using pairwise weights within block

    # copula-entropy knobs
    degree=2,
    max_num_mc=10_000,
    max_num_mc_adapt=0,
    alpha_reg=0.0,
    random_state=None,

    # stopping
    n_perm=1,
    stop_if_leq_perm_mean=False,
    min_gain=0.0,
):
    X = np.asarray(X)
    y = np.asarray(y)
    if y.ndim == 1:
        y = y.reshape(-1, 1)
    n, d = X.shape
    if max_features is None:
        max_features = d
    max_features = int(min(max_features, d))

    rng = np.random.default_rng(random_state)

    # -------- Copula entropy estimator --------
    from CopulaEntropy import CopulaEntropyEstimator
    ce = CopulaEntropyEstimator(
        degree=degree,
        max_num_mc=max_num_mc,
        max_num_mc_adapt=max_num_mc_adapt,
        alpha_reg=alpha_reg,
        random_state=random_state,
    )

    def Hc(Z):
        Z = np.asarray(Z)
        if Z.ndim == 1:
            Z = Z.reshape(-1, 1)
        h, _ = ce.compute(Z)
        return float(h)

    # -------- Rank map X once for structure learning (matches your framework) --------
    Ux = _modified_ecdf_to_uniform(X)

    # Pairwise dependence matrix on X (re-use your existing utility)
    W = _pairwise_dependence_graph(Ux, metric=dep_metric, top_k=dep_top_k)

    # -------- Learn global blocks via MST + cuts --------
    pa = _max_spanning_tree_parents(W, root=root)
    blocks, mst_edges, kept_mask = _blocks_from_mst(
        pa, W, min_edge_weight=min_edge_weight, max_block_size=max_block_size
    )

    # Map var -> block id
    var_block = np.full(d, -1, dtype=int)
    for b, idxs in enumerate(blocks):
        for j in idxs:
            var_block[int(j)] = b

    # reverse index for sparse updates: candidates in same block are affected
    rev = [[] for _ in range(d)]
    for j in range(d):
        bj = var_block[j]
        if bj < 0:
            continue
        for s in blocks[bj]:
            if int(s) != j:
                rev[int(s)].append(j)

    # -------- Block-factorized MI for any subset S of columns --------
    # Global block independence assumption:
    #   Hc(X_S) ≈ sum_b Hc(X_{S∩block_b})
    #   Hc([y, X_S]) via star around y across intersected blocks.
    hc_y = Hc(y)

    mi_cache_yS = {}  # cache I(y; S) where S is tuple of column indices (sorted)

    def mi_yS(cols):
        cols = tuple(sorted(cols))
        if cols in mi_cache_yS:
            return mi_cache_yS[cols]

        # per-block intersections
        hc_x = 0.0
        hc_yx_sum = 0.0
        B = 0

        for b_idxs in blocks:
            inter = [j for j in b_idxs if int(j) in cols]
            if not inter:
                continue
            B += 1
            Xb = X[:, inter]
            hc_xb = Hc(Xb)
            hc_x += hc_xb
            hc_yxb = Hc(np.concatenate([y, Xb], axis=1))
            hc_yx_sum += hc_yxb

        if B == 0:
            val = 0.0
        else:
            hc_yx = hc_yx_sum - (B - 1) * hc_y
            val = max(hc_x + hc_y - hc_yx, 0.0)

        mi_cache_yS[cols] = float(val)
        return float(val)

    def cond_gain(j, C):
        # I(y; xj | C) ≈ I(y; C∪{j}) - I(y; C)
        if len(C) == 0:
            return mi_yS((j,))
        base = mi_yS(C)
        full = mi_yS(tuple(sorted(C + (j,))))
        return max(float(full - base), 0.0)

    # -------- Seed: bivariate MI ranking --------
    mi_biv = np.empty(d, dtype=float)
    for j in range(d):
        mi_biv[j] = mi_yS((j,))
    first = int(np.argmax(mi_biv))

    selected = [first]
    gains = [float(mi_biv[first])]

    # -------- Candidate neighborhood definition using blocks --------
    selected_set = {first}
    cached_C = [None] * d
    cached_score = np.zeros(d, dtype=float)
    active = np.ones(d, dtype=bool)
    active[first] = False

    def make_C(j):
        bj = var_block[j]
        if bj < 0:
            return tuple()
        # only condition on selected features in the same block
        S = [s for s in blocks[bj] if int(s) in selected_set and int(s) != j]
        if len(S) <= sep_cap:
            return tuple(sorted(int(s) for s in S))
        # cap by pairwise weights
        S = [int(s) for s in S]
        S.sort(key=lambda s: W[j, s], reverse=True)
        return tuple(sorted(S[:sep_cap]))

    # initial scoring
    for j in range(d):
        if not active[j]:
            continue
        Cj = make_C(j)
        cached_C[j] = Cj
        cached_score[j] = cond_gain(j, Cj)

    # -------- Iterative selection --------
    while len(selected) < max_features:
        j_star = int(np.argmax(np.where(active, cached_score, -np.inf)))
        best = float(cached_score[j_star])
        if not np.isfinite(best) or best <= float(min_gain):
            break

        Cstar = cached_C[j_star]

        # permutation stop test (uses unadjusted score)
        if n_perm and stop_if_leq_perm_mean:
            best_unadj = cond_gain(j_star, Cstar)
            perm_gains = []
            xj = X[:, j_star].copy()
            for _ in range(int(n_perm)):
                X[:, j_star] = rng.permutation(xj)
                # invalidate cache entries that involve j_star
                # (cheap approach: just compute directly without caching)
                perm_gains.append(cond_gain(j_star, Cstar))
            X[:, j_star] = xj
            if best_unadj <= float(np.mean(perm_gains)):
                break

        # accept
        selected.append(j_star)
        gains.append(best)
        selected_set.add(j_star)
        active[j_star] = False

        # sparse update: only variables in same block can change C(j)
        for j in rev[j_star]:
            if not active[j]:
                continue
            newC = make_C(j)
            if newC == cached_C[j]:
                continue
            cached_C[j] = newC
            cached_score[j] = cond_gain(j, newC)

    return selected, gains, cached_score, {
        "blocks": blocks,
        "parents": pa,
        "mst_edges": mst_edges,
        "mst_kept_mask": kept_mask,
        "var_block": var_block,
    }
```

```python


import numpy as np

# -----------------------------------------------------------------------------
# Chordalization + maximal cliques (NEW)
# -----------------------------------------------------------------------------

def _mcs_order(adj: np.ndarray):
    """Maximum Cardinality Search. Returns an elimination order (list of vertices)."""
    d = adj.shape[0]
    w = np.zeros(d, dtype=np.int32)
    used = np.zeros(d, dtype=bool)
    order = []

    for _ in range(d):
        # pick unused vertex with max weight; tie-break by smallest index
        cand = np.where(~used)[0]
        v = cand[np.lexsort((cand, -w[cand]))[0]]
        used[v] = True
        order.append(int(v))
        nbrs = np.where(adj[v] & (~used))[0]
        w[nbrs] += 1

    return order  # first eliminated -> last eliminated


def _chordalize(adj: np.ndarray):
    """
    Chordalize (triangulate) an undirected graph using MCS ordering + fill edges.
    Returns:
      adj_fill: chordal supergraph adjacency (0/1 int32)
      peo: a perfect elimination order (last-to-first of MCS order is a PEO for chordalized graph)
      rank: rank[v] (higher rank = earlier in PEO direction used for clique extraction)
    """
    adj = (adj > 0).astype(np.int32)
    np.fill_diagonal(adj, 0)

    mcs = _mcs_order(adj.astype(bool))
    # For chordal completion, process vertices in reverse MCS to make it a PEO for the filled graph.
    peo = list(reversed(mcs))
    rank = np.empty(adj.shape[0], dtype=np.int32)
    for r, v in enumerate(peo):
        rank[v] = r  # smaller r = earlier in PEO

    adj_fill = adj.copy()

    # Fill edges: for each v, connect its higher-ranked neighbors into a clique.
    d = adj.shape[0]
    for v in peo:
        higher = np.where((adj_fill[v] == 1) & (rank > rank[v]))[0]
        if higher.size <= 1:
            continue
        # clique them up
        for i in range(higher.size):
            a = higher[i]
            row = higher[i + 1 :]
            if row.size == 0:
                break
            adj_fill[a, row] = 1
            adj_fill[row, a] = 1

    return adj_fill, peo, rank


def _maximal_cliques_from_chordal(adj_fill: np.ndarray, rank: np.ndarray):
    """
    In a chordal graph, maximal cliques can be generated from a PEO:
      C(v) = {v} ∪ N_+(v) where N_+(v) are neighbors with higher rank.
    Filter to maximal (remove subsets).
    """
    d = adj_fill.shape[0]
    cliques = []
    for v in range(d):
        higher = np.where((adj_fill[v] == 1) & (rank > rank[v]))[0]
        C = tuple(sorted(np.concatenate(([v], higher)).tolist()))
        if len(C) > 0:
            cliques.append(C)

    # keep only maximal cliques
    cliques = sorted(list(set(cliques)), key=lambda t: (len(t), t), reverse=True)
    maximal = []
    maximal_sets = []
    for C in cliques:
        Cs = set(C)
        is_subset = False
        for Ms in maximal_sets:
            if Cs.issubset(Ms):
                is_subset = True
                break
        if not is_subset:
            maximal.append(C)
            maximal_sets.append(Cs)
    return maximal


def _validate_running_intersection(cliques, jt_edges):
    """
    Running intersection property:
      for every variable x, cliques containing x must form a connected subtree in the clique tree.
    Returns (ok: bool, bad_vars: list[int]).
    """
    if len(cliques) <= 1:
        return True, []

    m = len(cliques)
    adjT = [[] for _ in range(m)]
    for i, j, _S in jt_edges:
        adjT[i].append(j)
        adjT[j].append(i)

    # map var -> clique indices that contain it
    var_to_cliques = {}
    for ci, C in enumerate(cliques):
        for v in C:
            var_to_cliques.setdefault(v, []).append(ci)

    bad = []
    for v, nodes in var_to_cliques.items():
        if len(nodes) <= 1:
            continue
        target = set(nodes)
        # BFS restricted to cliques that contain v
        start = nodes[0]
        q = [start]
        seen = {start}
        while q:
            a = q.pop()
            for b in adjT[a]:
                if b in seen:
                    continue
                if b in target:
                    seen.add(b)
                    q.append(b)
        if seen != target:
            bad.append(int(v))

    return (len(bad) == 0), bad


# -----------------------------------------------------------------------------
# Junction tree over cliques (unchanged: maximum overlap MST)
# -----------------------------------------------------------------------------


def _junction_tree_from_cliques(cliques):
    m = len(cliques)
    if m <= 1:
        return []

    edges = []
    for i in range(m):
        Ci = set(cliques[i])
        for j in range(i + 1, m):
            Cj = set(cliques[j])
            w = len(Ci & Cj)
            if w > 0:
                edges.append((w, i, j))

    if not edges:
        return []

    edges.sort(reverse=True)

    parent = list(range(m))
    rank = [0] * m

    def find(x):
        while parent[x] != x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x

    def union(a, b):
        ra, rb = find(a), find(b)
        if ra == rb:
            return False
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra] += 1
        return True

    jt_edges = []
    for w, i, j in edges:
        if union(i, j):
            sep = tuple(sorted(set(cliques[i]) & set(cliques[j])))
            jt_edges.append((i, j, sep))

    return jt_edges


import numpy as np

# -----------------------------------------------------------------------------
# Mutual information using the junction-tree copula-entropy approximation (FIXED)
# -----------------------------------------------------------------------------

def estimate_mi_junction_tree(
    X, Y,
    max_dim=5,
    min_dim=2,
    min_sep_dim=2,
    degree=2,
    num_mc=10000,
    dep_metric="mi",
    dep_top_k=None,
    random_state=None,
    solver="newton",
    qmc_engine="sobol",
    feature_mode="full",
    validate_running_intersection=False,
    fail_on_negative=True,
    _cache=None,
    _estimator=None,
    dep_mi_estimator=None,
    dep_mi_cache=None,
    dep_mi_degree=2,
    dep_mi_num_mc=2000,
    dep_mi_solver="newton",
    dep_mi_qmc_engine="sobol",
    dep_mi_feature_mode="pairwise",
):
    X = np.asarray(X)
    Y = np.asarray(Y)
    if X.ndim != 2 or Y.ndim != 2:
        raise ValueError("X and Y must be 2D arrays")
    if X.shape[0] != Y.shape[0]:
        raise ValueError("X and Y must have the same number of rows")

    XY = np.concatenate([X, Y], axis=1)

    h_xy = approx_copula_entropy_junction_tree(
        XY,
        max_dim=max_dim, min_dim=min_dim, min_sep_dim=min_sep_dim,
        degree=degree, num_mc=num_mc,
        dep_metric=dep_metric, dep_top_k=dep_top_k,
        random_state=random_state, solver=solver, qmc_engine=qmc_engine,
        feature_mode=feature_mode,
        validate_running_intersection=validate_running_intersection,
        fail_on_negative=fail_on_negative,
        _cache=_cache, _estimator=_estimator,
        dep_mi_estimator=dep_mi_estimator, dep_mi_cache=dep_mi_cache,
        dep_mi_degree=dep_mi_degree, dep_mi_num_mc=dep_mi_num_mc,
        dep_mi_solver=dep_mi_solver, dep_mi_qmc_engine=dep_mi_qmc_engine,
        dep_mi_feature_mode=dep_mi_feature_mode,
    )

    h_x = approx_copula_entropy_junction_tree(
        X,
        max_dim=max_dim, min_dim=min_dim, min_sep_dim=min_sep_dim,
        degree=degree, num_mc=num_mc,
        dep_metric=dep_metric, dep_top_k=dep_top_k,
        random_state=random_state, solver=solver, qmc_engine=qmc_engine,
        feature_mode=feature_mode,
        validate_running_intersection=validate_running_intersection,
        fail_on_negative=fail_on_negative,
        _cache=_cache, _estimator=_estimator,
        dep_mi_estimator=dep_mi_estimator, dep_mi_cache=dep_mi_cache,
        dep_mi_degree=dep_mi_degree, dep_mi_num_mc=dep_mi_num_mc,
        dep_mi_solver=dep_mi_solver, dep_mi_qmc_engine=dep_mi_qmc_engine,
        dep_mi_feature_mode=dep_mi_feature_mode,
    )

    h_y = approx_copula_entropy_junction_tree(
        Y,
        max_dim=max_dim, min_dim=min_dim, min_sep_dim=min_sep_dim,
        degree=degree, num_mc=num_mc,
        dep_metric=dep_metric, dep_top_k=dep_top_k,
        random_state=random_state, solver=solver, qmc_engine=qmc_engine,
        feature_mode=feature_mode,
        validate_running_intersection=validate_running_intersection,
        fail_on_negative=fail_on_negative,
        _cache=_cache, _estimator=_estimator,
        dep_mi_estimator=dep_mi_estimator, dep_mi_cache=dep_mi_cache,
        dep_mi_degree=dep_mi_degree, dep_mi_num_mc=dep_mi_num_mc,
        dep_mi_solver=dep_mi_solver, dep_mi_qmc_engine=dep_mi_qmc_engine,
        dep_mi_feature_mode=dep_mi_feature_mode,
    )

    mi = float(h_x + h_y - h_xy)
    return max(mi, 0.0)



# -----------------------------------------------------------------------------
# Copula entropy via junction-tree-like clique/separator sum (UPDATED)
#   - chordalize learned graph (MUST)
#   - use maximal cliques of chordal graph
#   - optionally validate running intersection
#   - monitor failures: if result < 0 -> invalid structure / inconsistent entropies
# -----------------------------------------------------------------------------
def approx_copula_entropy_junction_tree(
    samples,
    max_dim=5,
    min_dim=2,
    min_sep_dim=2,
    degree=2,
    num_mc=10000,
    dep_metric="mi",          # CHANGED default
    dep_top_k=None,
    random_state=None,
    solver="newton",
    qmc_engine="sobol",
    feature_mode="full",
    undersized="merge",
    _cache=None,
    _estimator=None,
    validate_running_intersection=False,
    fail_on_negative=True,
    # NEW: MI-based pairwise graph learning controls
    dep_mi_estimator=None,
    dep_mi_cache=None,
    dep_mi_degree=2,
    dep_mi_num_mc=2000,
    dep_mi_solver="newton",
    dep_mi_qmc_engine="sobol",
    dep_mi_feature_mode="pairwise",
):
    n, d = samples.shape

    if min_dim < 1:
        raise ValueError("min_dim must be >= 1")
    if max_dim < min_dim:
        raise ValueError("max_dim must be >= min_dim")
    if d < min_dim:
        return 0.0

    if _cache is None:
        _cache = {}

    if _estimator is None:
        _estimator = CopulaEntropyEstimator(
            degree=degree,
            num_mc=num_mc,
            random_state=random_state,
            qmc_engine=qmc_engine,
            solver=solver,
            feature_mode=feature_mode,
        )

    def entropy_of(cols):
        cols = tuple(sorted(cols))
        k = len(cols)
        if k < min_dim:
            return 0.0
        if k > max_dim:
            cols = cols[:max_dim]
            k = len(cols)
        key = (cols, degree, num_mc, solver, qmc_engine, feature_mode, random_state)
        if key in _cache:
            return _cache[key]
        sub = samples[:, cols]
        h, _ = _estimator.compute(sub)
        _cache[key] = h
        return h

    # If small enough, estimate directly
    if d <= max_dim:
        return float(entropy_of(tuple(range(d))))

    # Learn weighted graph via bivariate mutual information (CHANGED)
    W = _pairwise_dependence_graph(
    )

    # Build unweighted adjacency from learned weights
    adj = (W > 0).astype(np.int32)
    np.fill_diagonal(adj, 0)

    # 1) chordalize (MUST)
    adj_fill, peo, rank = _chordalize(adj)

    # 2) maximal cliques of chordal graph (MUST)
    cliques = _maximal_cliques_from_chordal(adj_fill, rank)
    # drop too-small cliques (still maximal, but not useful for estimator)
    cliques = [C for C in cliques if len(C) >= min_dim]

    # 3) clique tree by maximum overlap (what you do)
    jt_edges = _junction_tree_from_cliques(cliques)

    # 4) optional validation: running intersection
    if validate_running_intersection:
        ok, bad_vars = _validate_running_intersection(cliques, jt_edges)
        if not ok:
            raise RuntimeError(
                f"Running-intersection violated for variables: {bad_vars}. "
                f"Chordalization/clique extraction or clique-tree construction is inconsistent."
            )

    # Clique sum
    h_cliques = 0.0
    for C in cliques:
        h_cliques += entropy_of(C)

    # Separator sum
    effective_min_sep = max(min_sep_dim, min_dim)
    h_seps = 0.0
    for i, j, S in jt_edges:
        if len(S) < effective_min_sep:
            continue
        h_seps += entropy_of(S)

    out = float(h_cliques - h_seps)

    # 5) monitor failures: negative result
    if out < 0.0:
        msg = (
            f"JT entropy estimate < 0 ({out}). "
            f"Structure may be invalid or entropy estimates inconsistent."
        )
        if fail_on_negative:
            raise RuntimeError(msg)
        return 0.0

    return out

# Optional: expose learned structure (PATCHED)
#   - default dep_metric to 'mi'
# -----------------------------------------------------------------------------


def learn_jt_structure_from_samples(
    U,
    max_dim=5,
    min_dim=2,
    dep_metric="mi",          # CHANGED default
    dep_top_k=None,
    undersized="merge",
):
    W = _pairwise_dependence_graph(U, metric=dep_metric, top_k=dep_top_k)
    adj = (W > 0).astype(np.int32)
    np.fill_diagonal(adj, 0)

    adj_fill, peo, rank = _chordalize(adj)
    cliques = _maximal_cliques_from_chordal(adj_fill, rank)
    cliques = [C for C in cliques if len(C) >= min_dim]
    jt_edges = _junction_tree_from_cliques(cliques)
    return cliques, jt_edges


# -----------------------------------------------------------------------------
# Pairwise dependency graph (REPLACED)
#   - MI only: bivariate mutual information via CopulaEntropyEstimator on 2D
#   - MI = max(-H_copula(U_i,U_j), 0)
# -----------------------------------------------------------------------------


def _pairwise_dependence_graph(
    U,
    metric="mi",
    top_k=None,
    mi_degree=2,
    mi_num_mc=2000,
    mi_solver="newton",
    mi_qmc_engine="sobol",
    mi_feature_mode="pairwise",
    _mi_estimator=None,
    _mi_cache=None,
    random_state=None,
):
    n, d = U.shape
    W = np.zeros((d, d), dtype=float)



    if _mi_cache is None:
        _mi_cache = {}

    if _mi_estimator is None:
        _mi_estimator = CopulaEntropyEstimator(
            degree=mi_degree,
            max_num_mc=mi_num_mc,
            random_state=random_state,

            feature_mode=mi_feature_mode,
        )

    for i in range(d):
        for j in range(i + 1, d):
            key = (i, j, mi_degree, mi_num_mc, mi_solver, mi_qmc_engine, mi_feature_mode, random_state)
            if key in _mi_cache:
                mij = _mi_cache[key]
            else:
                h, _ = _mi_estimator.compute(U[:, (i, j)])
                mij = float(max(-h, 0.0))
                _mi_cache[key] = mij
            W[i, j] = W[j, i] = mij

    np.fill_diagonal(W, 0.0)

    if top_k is not None and top_k < d - 1:
        W_sparse = np.zeros_like(W)
        for i in range(d):
            idx = np.argsort(W[i])[::-1]
            keep = idx[:top_k]
            W_sparse[i, keep] = W[i, keep]
        W = np.maximum(W_sparse, W_sparse.T)
        np.fill_diagonal(W, 0.0)

    return W
```

```python
# pip install pmlb lightgbm scikit-learn pandas numpy

from __future__ import annotations

import time
import json
from dataclasses import dataclass, asdict

import numpy as np
import pandas as pd

from pmlb import fetch_data, classification_dataset_names, regression_dataset_names
from lightgbm import LGBMClassifier, LGBMRegressor

from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import balanced_accuracy_score, r2_score

# -------------------------------------------------------------------
# ASSUMPTION:
# bf_separator_forward_select(...) is already importable / defined.
# It also depends on your internal helpers (_pairwise_dependence_graph,
# _max_spanning_tree_parents, _blocks_from_mst, etc.) being available.
# -------------------------------------------------------------------
# from your_module import bf_separator_forward_select


RANDOM_STATE = 42
N_SPLITS = 5

# Selection settings for your function
SELECTOR_KWARGS = dict(
    max_features=9999,          # cap selected subset size for tractability
    dep_metric="spearman",
    dep_top_k=None,
    root=0,
    min_edge_weight=0.0,
    max_block_size=6,
    sep_cap=6,
    degree=2,
    max_num_mc=3_000,         # lower than defaults for benchmark speed
    max_num_mc_adapt=1_000,
    alpha_reg=0.0,
    random_state=RANDOM_STATE,
    n_perm=0,
    stop_if_leq_perm_mean=True,
    min_gain=0.0,
)

# LightGBM settings
LGBM_CLASSIFIER_KWARGS = dict(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbosity=-1,
)

LGBM_REGRESSOR_KWARGS = dict(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbosity=-1,
)


@dataclass
class FoldResult:
    task: str
    dataset: str
    fold: int
    n_samples: int
    n_features: int
    n_selected: int
    full_score: float
    selected_score: float
    permuted_selected_score: float
    selected_minus_full: float
    selected_minus_permuted: float
    fit_seconds_selector: float
    selected_features: list[int]


def _ensure_numeric_2d(X):
    if isinstance(X, pd.DataFrame):
        X = X.to_numpy()
    X = np.asarray(X)
    if X.ndim != 2:
        raise ValueError(f"Expected 2D X, got {X.ndim}D")
    return X.astype(np.float64, copy=False)


def _encode_target_classification(y):
    if isinstance(y, (pd.Series, pd.DataFrame)):
        y = np.asarray(y).ravel()
    else:
        y = np.asarray(y).ravel()

    le = LabelEncoder()
    y_enc = le.fit_transform(y)
    return y_enc


def _encode_target_regression(y):
    if isinstance(y, (pd.Series, pd.DataFrame)):
        y = np.asarray(y).ravel()
    else:
        y = np.asarray(y).ravel()
    return y.astype(np.float64, copy=False)


def _permute_columns(X, rng):
    Xp = np.array(X, copy=True)
    n_cols = Xp.shape[1]
    for j in range(n_cols):
        Xp[:, j] = rng.permutation(Xp[:, j])
    return Xp


def _score_classification(y_true, y_pred):
    return balanced_accuracy_score(y_true, y_pred)


def _score_regression(y_true, y_pred):
    return r2_score(y_true, y_pred)


def _make_cv(task, y):
    if task == "classification":
        return StratifiedKFold(
            n_splits=N_SPLITS,
            shuffle=True,
            random_state=RANDOM_STATE,
        )
    return KFold(
        n_splits=N_SPLITS,
        shuffle=True,
        random_state=RANDOM_STATE,
    )


def _make_model(task):
    if task == "classification":
        return LGBMClassifier(**LGBM_CLASSIFIER_KWARGS)
    return LGBMRegressor(**LGBM_REGRESSOR_KWARGS)


def _fit_eval_model(task, X_train, y_train, X_test, y_test):
    model = _make_model(task)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    if task == "classification":
        return _score_classification(y_test, pred)
    return _score_regression(y_test, pred)


def _select_features(X_train, y_train):
    t0 = time.perf_counter()
    selected, gains, cached_score, meta = bf_separator_forward_select(
        X_train,
        y_train,
        **SELECTOR_KWARGS,
    )
    dt = time.perf_counter() - t0

    selected = np.asarray(selected, dtype=int)
    if selected.size == 0:
        # fallback: if your selector returns nothing, keep the best univariate
        selected = np.array([0], dtype=int)

    return selected, dt, gains, cached_score, meta


def benchmark_one_dataset(task, dataset_name):
    X, y = fetch_data(dataset_name, return_X_y=True)
    X = _ensure_numeric_2d(X)

    print(X.shape)

    if task == "classification":
        y = _encode_target_classification(y)
    else:
        y = _encode_target_regression(y)

    n_samples, n_features = X.shape

    if not (n_features > 12 and n_samples > 400 and n_samples < 250_000):
        return []

    cv = _make_cv(task, y)
    results = []

    for fold, (tr_idx, te_idx) in enumerate(cv.split(X, y), start=1):
        X_train, X_test = X[tr_idx], X[te_idx]
        y_train, y_test = y[tr_idx], y[te_idx]

        # 1) Full features
        full_score = _fit_eval_model(task, X_train, y_train, X_test, y_test)

        # 2) Selected features
        selected, select_seconds, gains, cached_score, meta = _select_features(X_train, y_train)
        X_train_sel = X_train[:, selected]
        X_test_sel = X_test[:, selected]
        selected_score = _fit_eval_model(task, X_train_sel, y_train, X_test_sel, y_test)

        # 3) Permuted-selected baseline
        rng = np.random.default_rng(RANDOM_STATE + fold)
        X_train_perm = _permute_columns(X_train_sel, rng)
        X_test_perm = _permute_columns(X_test_sel, rng)
        permuted_selected_score = _fit_eval_model(task, X_train_perm, y_train, X_test_perm, y_test)

        results.append(
            FoldResult(
                task=task,
                dataset=dataset_name,
                fold=fold,
                n_samples=n_samples,
                n_features=n_features,
                n_selected=int(len(selected)),
                full_score=float(full_score),
                selected_score=float(selected_score),
                permuted_selected_score=float(permuted_selected_score),
                selected_minus_full=float(selected_score - full_score),
                selected_minus_permuted=float(selected_score - permuted_selected_score),
                fit_seconds_selector=float(select_seconds),
                selected_features=selected.tolist(),
            )
        )

    return results


def benchmark_pmlb(
    max_classification_datasets=None,
    max_regression_datasets=None,
    save_prefix="bf_selector_pmlb",
):
    all_results = []

    cls_names = list(classification_dataset_names)
    reg_names = list(regression_dataset_names)

    if max_classification_datasets is not None:
        cls_names = cls_names[:max_classification_datasets]
    if max_regression_datasets is not None:
        reg_names = reg_names[:max_regression_datasets]

    for task, names in [
        ("classification", cls_names),
        ("regression", reg_names),
    ]:
        for name in names:
            print(f"[{task}] {name}")
            ds_results = benchmark_one_dataset(task, name)
            all_results.extend(ds_results)

    df = pd.DataFrame([asdict(r) for r in all_results])

    if df.empty:
        print("No datasets passed the filter.")
        return df, pd.DataFrame()

    summary = (
        df.groupby(["task", "dataset"], as_index=False)
        .agg(
            n_samples=("n_samples", "first"),
            n_features=("n_features", "first"),
            mean_n_selected=("n_selected", "mean"),
            mean_full_score=("full_score", "mean"),
            mean_selected_score=("selected_score", "mean"),
            mean_permuted_selected_score=("permuted_selected_score", "mean"),
            mean_selected_minus_full=("selected_minus_full", "mean"),
            mean_selected_minus_permuted=("selected_minus_permuted", "mean"),
            mean_selector_seconds=("fit_seconds_selector", "mean"),
        )
        .sort_values(["task", "mean_selected_minus_permuted"], ascending=[True, False])
        .reset_index(drop=True)
    )

    overall = (
        summary.groupby("task", as_index=False)
        .agg(
            datasets=("dataset", "count"),
            avg_full_score=("mean_full_score", "mean"),
            avg_selected_score=("mean_selected_score", "mean"),
            avg_permuted_selected_score=("mean_permuted_selected_score", "mean"),
            avg_selected_minus_full=("mean_selected_minus_full", "mean"),
            avg_selected_minus_permuted=("mean_selected_minus_permuted", "mean"),
            avg_n_selected=("mean_n_selected", "mean"),
            avg_selector_seconds=("mean_selector_seconds", "mean"),
        )
    )

    df.to_csv(f"{save_prefix}_fold_results.csv", index=False)
    summary.to_csv(f"{save_prefix}_dataset_summary.csv", index=False)
    overall.to_csv(f"{save_prefix}_overall_summary.csv", index=False)

    print("\n=== OVERALL ===")
    print(overall.to_string(index=False))

    print("\n=== TOP DATASETS BY selected - permuted ===")
    print(
        summary.sort_values(
            ["task", "mean_selected_minus_permuted"],
            ascending=[True, False],
        )
        .groupby("task")
        .head(15)
        .to_string(index=False)
    )

    return df, summary


if __name__ == "__main__":
    fold_df, summary_df = benchmark_pmlb(
        max_classification_datasets=None,  # set an int for a quick smoke test
        max_regression_datasets=None,      # set an int for a quick smoke test
        save_prefix="bf_selector_pmlb",
    )
```
